```python
df=pd.read_csv('/kaggle/input/second-hand-car-price-prediction/cars.csv')
df.head()
```

```python
df.info()
```

```python
df.drop("Car_ID",axis=1,inplace=True)
df
```

```python
df.drop("Year",axis=1,inplace=True)
df
```

```python
df.describe()
```

```python
Y = df['Price']
df.drop("Price", axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Fuel_Type_num']=pd.get_dummies(df.Fuel_Type,drop_first=True,dtype=int)
df
```

```python
df.drop("Fuel_Type",axis=1,inplace=True)
df
```

```python
df['Transmission_num']=pd.get_dummies(df.Transmission,drop_first=True,dtype=int)
df
```

```python
df.drop("Transmission",axis=1,inplace=True)
df
```

```python
dummies=pd.get_dummies(df["Brand"],drop_first=True,dtype=int)
dummies
```

```python
df=pd.concat([df,dummies],axis=1)
df
```

```python
df.drop('Brand',inplace=True,axis=1)
df
```

```python
model_dummies=pd.get_dummies(df['Model'],drop_first=True,dtype=int)
model_dummies
```

```python
df=pd.concat([df,model_dummies],axis=1)
df
```

```python
df.drop('Model',inplace=True,axis=1)
df
```

```python
Owner_dummy=pd.get_dummies(df['Owner_Type'],drop_first=True,dtype=int)
Owner_dummy
```

```python
df=pd.concat([df,Owner_dummy],axis=1)
df
```

```python
df.drop('Owner_Type',axis=1,inplace=True)
df
```

```python

X=df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

# RMSE->root mean squared root errors

```python
predictions=model.predict(X_test)
```

# applying decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=3,min_samples_split=10 ,random_state=42)
model.fit(X_train,Y_train)
print(model.score(X_train,Y_train))
print(model.score(X_test,Y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(10,bootstrap=False,max_depth=3)
model.fit(X_train,Y_train)
print(model.score(X_train,Y_train))
print(model.score(X_test,Y_test))
```
